```python
from functools import partial
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
import nbody
```

```python
masses, initial_values = nbody.load_bodies_from_json('pythag.json')

d = partial(nbody.derivatives, masses)
sol = solve_ivp(d, [0, 100], initial_values, max_step=.001)
```

```python
nbody.draw_bodies(masses, sol.t, sol.y, animate=False)
```

```python
nbody.draw_stats(masses, sol.t, sol.y)
```
